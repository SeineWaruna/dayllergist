```python
import pandas as pd

df_deepseekchat = pd.read_csv("dayllergist_baseline_deepseekchat.csv")
df_deepseekreasoner = pd.read_csv("dayllergist_baseline_deepseekreasoner.csv")
df_gemini25 = pd.read_csv("dayllergist_baseline_gemini25.csv")
df_gpt4omini = pd.read_csv("dayllergist_baseline_gpt4omini.csv")
df_gpt4turbo = pd.read_csv("dayllergist_baseline_gpt4turbo.csv")
```

```python
sample1 = df_deepseekchat.sample(n=20, random_state=42).copy()
sample1["Model"] = "deepseekchat"

sample2 = df_deepseekreasoner.sample(n=20, random_state=42).copy()
sample2["Model"] = "deepseekreasoner"

sample3 = df_gemini25.sample(n=20, random_state=42).copy()
sample3["Model"] = "gemini25"

sample4 = df_gpt4omini.sample(n=20, random_state=42).copy()
sample4["Model"] = "gpt4omini"

sample5 = df_gpt4turbo.sample(n=20, random_state=42).copy()
sample5["Model"] = "gpt4turbo"
```

```python
combined_df = pd.concat([sample1, sample2, sample3, sample4, sample5], ignore_index=True)
shuffle_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)
shuffle_df.to_csv("combined_shuffle_100.csv", index=False)
```

```python
df1 = pd.read_csv("dayllergist_baseline_deepseekchat.csv").iloc[:24].copy()
df2 = pd.read_csv("dayllergist_baseline_deepseekreasoner.csv").iloc[:24].copy()
df3 = pd.read_csv("dayllergist_baseline_gemini25.csv").iloc[:24].copy()
df4 = pd.read_csv("dayllergist_baseline_gpt4omini.csv").iloc[:24].copy()
df5 = pd.read_csv("dayllergist_baseline_gpt4turbo.csv").iloc[:24].copy()
```

```python
df1["Model"] = "deepseekchat"
df2["Model"] = "deepseekreasoner"
df3["Model"] = "gemini25"
df4["Model"] = "gpt4omini"
df5["Model"] = "gpt4turbo"
```

```python
models = [df1, df2, df3, df4, df5]
```

```python
final_rows = []
for i in range(120):
    model_index = i % 5
    row_index = i // 5
    final_rows.append(models[model_index].iloc[row_index])
```

```python
final_df = pd.DataFrame(final_rows)
final_df.to_csv("interleaved_120_by_pattern.csv", index=False)
```

```python
import pandas as pd

# Load all 5 uploaded CSV files
file_paths = [
    "dayllergist_baseline_deepseekchat_1.csv",
    "dayllergist_baseline_deepseekreasoner.csv",
    "dayllergist_baseline_gemini25.csv",
    "dayllergist_baseline_gpt4omini.csv",
    "dayllergist_baseline_gpt4turbo.csv"
]

# Read the files into dataframes
dfs = [pd.read_csv(path) for path in file_paths]

# Ensure each file has at least 24 entries
for i, df in enumerate(dfs):
    assert len(df) >= 24, f"File {i+1} has fewer than 24 entries."

# Initialize a list for the final combined dataset
combined_data = []

# Collect 24 entries in an alternating pattern across the 5 files
for i in range(24):
    for j in range(5):
        entry_index = i * 5 + j
        combined_data.append(dfs[j].iloc[i])

# Create a new DataFrame from the combined data
combined_df = pd.DataFrame(combined_data)

# Display the result
# import ace_tools as tools; 
# tools.display_dataframe_to_user(name="Combined Alternating Dataset", dataframe=combined_df)
```
