```python
from google import genai
import pandas as pd
import time

client = genai.Client(api_key="AIzaSyBVon-UJpRDZTPkMtaXneG2JWLR4ASOLBw")

context = (
    "Adopt the role of an experienced allergist with extensive knowledge of eye and nose allergies. "
    "Provide accurate, in-depth analyses, and communicate in a professional, doctor-like tone, offering clear and expert advice. "
    "Focus on concise and short answers. The answers should be 2 sentences maximum."
)

if 'dayllergist_dataset_response' not in globals():
    dayllergist_dataset_response = []

for i in range(len(dayllergist_dataset_response), len(dayllergist_dataset_request)):
    question = dayllergist_dataset_request[i]
    prompt = f"{context}\n\nQ: {question}"

    try:
        response = client.models.generate_content(
            model="models/gemini-1.5-flash-latest",
            contents=prompt
        )
        completion_content = response.text.strip()
        if not completion_content:
            completion_content = "The response is empty!"
    except Exception as e:
        completion_content = f"Error: {str(e)}"

    print(f"[{i+1}/{len(dayllergist_dataset_request)}] {completion_content}")
    dayllergist_dataset_response.append(completion_content)
    time.sleep(1)

while len(dayllergist_dataset_response) < len(dayllergist_dataset_request):
    dayllergist_dataset_response.append("ERROR: No response")

dayllergist_dataset_gemini = pd.DataFrame({
    "request": dayllergist_dataset_request,
    "response": dayllergist_dataset_response
})

dayllergist_dataset_gemini.to_csv("dayllergist_baseline_gemini25.csv", index=False)
```
